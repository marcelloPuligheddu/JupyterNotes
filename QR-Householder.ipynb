```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

n, m = 12, 12
A = np.random.random((n,m))*2-1
np.set_printoptions(precision=2,suppress=True,linewidth=120)
A
```

```python
Ap = A.copy()
Q = np.eye(n)

## Householder applied recursevily to the (n-k).(n-k) submatrix
## Assumes A is **very** well behaved
for k in range(n-1):
    x = Ap[k:,k]
    e = np.zeros_like(x); e[0] = 1.
    a = - np.sign(x[0]) * np.sqrt(np.sum(x*x)) # n-k
    u = x + a * e                              # n-k
    v = u / np.sqrt(np.sum(u*u))               # n-k
    Qp = np.eye(n)                             #
    Qp[k:,k:] += - 2. * np.outer(v , v.T)      # 
    Ap = Qp@Ap                                 # should be the larf routine from lapack -> (n-k+1)**2
    Q = Qp@Q                                   # should be the larf routine from lapack -> (n-k+1)**2
Q = Q.T    
#    print(k,Ap)

R = (Q.T@A)

print(A)
print('=')
print(Q)
print('@')
print(R)
print()
print(' A-QR = ', np.max(A-Q@R))
```
